```python
import pandas as pd
import matplotlib as plt
```

```python
#read csv and print dataframe
csv = "Crime_Data_from_2020_to_Present.csv"
crime_data = pd.read_csv(csv)
crime_data.head()
```

```python
#remove miscellaneous columns
clean_crime_data = crime_data.drop(['DR_NO',
                                   'AREA',
                                   'Rpt Dist No',
                                   'Part 1-2',
                                   'Crm Cd Desc',
                                   'Status',
                                   'Status Desc',
                                   'Crm Cd 1',
                                   'Crm Cd 2',
                                   'Crm Cd 3',
                                   'Crm Cd 4',
                                   'LOCATION',
                                    'Cross Street',
                                   'Premis Cd',
                                   'Premis Desc',
                                   'Weapon Used Cd',
                                   'Weapon Desc',
                                   'Mocodes',
                                   'Date Rptd'], 
                                   axis=1)
clean_crime_data.head()
```

```python
#extract month and year from "DATE OCC"
clean_crime_data["Year of Occurrence"] = pd.DatetimeIndex(clean_crime_data["DATE OCC"]).year
clean_crime_data["Month of Occurrence"] = pd.DatetimeIndex(clean_crime_data["DATE OCC"]).month
clean_crime_data.head()
```

```python
#remove 2023 data from dataset
crime_2020_to_2022 = clean_crime_data.loc[clean_crime_data["Year of Occurrence"] != 2023]
crime_2020_to_2022.head()
```
